```python
import re
import os

# Check current working directory
print(f"Current working directory: {os.getcwd()}")

# Set data folder path
data_folder = "../data"

# 1. LOAD: Read the file from Step 1
input_file = os.path.join(data_folder, "1_raw_transcript.txt")
try:
    with open(input_file, "r") as f:
        raw_text = f.read()
    print(f"Successfully loaded {input_file} ({len(raw_text)} characters)")
except FileNotFoundError as e:
    print(f"Error: {e}")
    raw_text = None

if raw_text:
    # 2. PROCESS: Your cleaning function
    def clean_text(text):
        text = text.lower()
        text = re.sub(r'\b(um|uh|hmm|ah|like)\b', '', text)
        return re.sub(r'\s+', ' ', text).strip()

    cleaned_text = clean_text(raw_text)

    # 3. SAVE: Write to a new file for the next script
    output_file = os.path.join(data_folder, "2_cleaned_transcript.txt")
    with open(output_file, "w") as f:
        f.write(cleaned_text)

    print(f"Step 2 Complete: {output_file} created.")
```
